```python
# Marina Pier Inc. is leveraging technology to automate their operations on the San Francisco port.
# The company’s management has set out to build a bias-free/ corruption-free automatic system that reports & avoids faulty situations caused by human error. Examples of human error include misclassifying the correct type of boat. The type of boat that enters the port region is as follows.
# ●	Buoy
# ●	Cruise_ship
# ●	Ferry_boat
# ●	Freight_boar
# ●	Gondola
# ●	Inflatable_boat
# ●	Kayak
# ●	Paper_boat
# ●	Sailboat
# Marina Pier wants to use Deep Learning techniques to build an automatic reporting system that recognizes the boat. The company is also looking to use a transfer learning approach of any lightweight pre-trained model in order to deploy in mobile devices.
```

```python
import tensorflow as tf
import os
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.preprocessing.image import ImageDataGenerator

dataset_path = 'boat_type_classification_dataset'

# List the contents of the directory
contents = os.listdir(dataset_path)

# Print the contents
print("Contents of the directory:")
for item in contents:
    print(item)
    
```

```python
# Dataset and Data Description: 
# boat_type_classification_dataset.zip
# The dataset contains images of 9 types of boats. It contains a total of 1162 images. The training images are provided in the directory of the specific class itself. 
# Classes:
# ●	ferry_boat
# ●	gondola
# ●	sailboat
# ●	cruise_ship
# ●	kayak
# ●	inflatable_boat
# ●	paper_boat
# ●	buoy
# ●	freight_boat
```

```python
# As a deep learning engineer, your task is to:
# 1.	Build a CNN network to classify the boat.
# 2.	Build a lightweight model with the aim of deploying the solution on a mobile device using transfer learning. You can use any lightweight pre-trained model as the initial (first) layer. MobileNetV2 is a popular lightweight pre-trained model built using Keras API. 
```

```python
# When you use image_dataset_from_directory, Keras automatically encodes your string labels into integers based on their alphabetical order.


# Load the training dataset
train_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    dataset_path,
    image_size=(224, 224),  # Resize images to 256x256
    batch_size=32,  # Number of images to process in a batch
    validation_split=0.2,  # Reserve 20% of the data for validation
    subset="training",  # Specify this subset as the training data
    seed=123  # Seed for reproducibility
)

# Load the validation dataset
val_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    dataset_path,
    image_size=(224, 224),
    batch_size=32,
    validation_split=0.2,  # Use the same validation split
    subset="validation",  # Specify this subset as the validation data
    seed=123  # Same seed for consistency
)

# Optionally, print class names to verify
class_names = train_dataset.class_names
print("Class names:", class_names)
```

```python
class_names = train_dataset.class_names
print(class_names)
```

```python
# Visualize the data

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
for images, labels in train_dataset.take(1):
      for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
```

```python
# Normalize the pixel values (scale them to the range [0, 1])
image_scale = 1. / 255

train_dataset = train_dataset.map(lambda x, y: (x * image_scale, y))
val_dataset = val_dataset.map(lambda x, y: (x * image_scale, y))
```

```python
from tensorflow.keras import models, layers

# Define the CNN model
def create_cnn_model(input_shape, num_classes):
    model = models.Sequential()

    # Convolutional layers
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(128, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))

    # Flatten layer to transition from convolutional layers to fully connected layers
    model.add(layers.Flatten())

    # Fully connected layers
    model.add(layers.Dense(128, activation='relu'))
    model.add(layers.Dense(num_classes, activation='softmax'))

    return model

# Define input shape and number of classes
input_shape = (224, 224, 3)  # Assuming input images are resized to 224x224 RGB images
num_classes = 9  # Number of boat classes

# Create the CNN model
cnn_model = create_cnn_model(input_shape, num_classes)
```

```python
# Compile the model
cnn_model.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss='sparse_categorical_crossentropy',  # Use for integer labels
    metrics=[tf.keras.metrics.SparseCategoricalAccuracy(name='accuracy')]
)

# Train the model
cnn_model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=10  # Adjust the number of epochs as necessary
)
```

```python
# Evaluate the model on the validation dataset
loss, accuracy = cnn_model.evaluate(val_dataset)  # Use test_dataset if you have one

print(f"Validation Loss: {loss}")
print(f"Validation Accuracy: {accuracy}")
```

```python
# Validation Loss: 3.6292 suggests that the model is not fitting the data well. A high loss indicates that the predictions are far from the true labels.
# Validation Accuracy: 0.4397 (or around 44%) shows that the model is correctly classifying only about 44% of the validation samples, which is relatively low.
```

```python
# Next Steps to Improve Model Performance
# Data Augmentation:

# To help the model generalize better, you can augment your training data by applying transformations such as rotation, flipping, and zooming.
```

```python
# Define data augmentation parameters
train_datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Load and augment training data
train_generator = train_datagen.flow_from_directory(
    dataset_path,
    target_size=(224, 224),
    batch_size=32,  # Using a batch size greater than 1
    class_mode='categorical',
    shuffle=True
)

# Load testing data validation set was not augmanted 
test_datagen = ImageDataGenerator()
test_generator = test_datagen.flow_from_directory(
    dataset_path,
    target_size=(224, 224),
    batch_size=32,  # Using a batch size greater than 1
    class_mode='categorical',
    shuffle=True
)
# “The larger the batch size, the faster the model per epoch during training.”
```

```python
# Display both original and augmented images
num_images_to_display = 4

plt.figure(figsize=(12, 6))
for i in range(num_images_to_display):
    # Get an image from the generator
    image, _ = train_generator.next()

    # Plot original image
    plt.subplot(2, num_images_to_display, i + 1)
    plt.imshow(image.squeeze() / 255.0)  # Normalize pixel values to [0, 1]
    plt.title('Original')
    plt.axis('off')

    # Plot augmented image
    augmented_image, _ = train_generator.next()
    plt.subplot(2, num_images_to_display, num_images_to_display + i + 1)
    plt.imshow(augmented_image.squeeze() / 255.0)
    plt.title('Augmented')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Begin with fitting the augmented traning data and check the valiation accuracy ,
# traing accuracy, validaton loss and traing loss and cimpare to the model that trained with no augmentation data  
```

```python
# Create the CNN model
cnn_model_augmented = create_cnn_model(input_shape, num_classes)
```

```python
# Compile the model
cnn_model_augmented.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# Fit the model
history_augmented = cnn_model_augmented.fit(
    train_generator,
    validation_data=test_generator,
    epochs=20,
    callbacks=[tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)]
)
```

```python
# Training Loss: indicating that the model is learning from the training data.
# Validation Loss: Ideally, this should decrease alongside training loss; if it starts to increase while training loss decreases, it may indicate overfitting.
```

```python
# Training Accuracy: The training accuracy has reached approximately 40.10%, showing that the model has improved its ability to classify the training data correctly over the epochs.
# Validation Accuracy: The validation accuracy is at 43.20%, which indicates that the model has also improved in generalizing to unseen data during validation.
```

```python
# Training Loss: The training loss has decreased to 1.6807, which is a positive sign as it suggests the model is learning effectively.
# Validation Loss: The validation loss of 1.6440 is close to the training loss, which can indicate that the model is performing well on both the training and validation datasets.
```
